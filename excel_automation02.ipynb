```python
# 問題２　ファイルの集計自動化コードを作成しよう
# globモジュールを使い'branch_offices'フォルダにある、拡張子がxlsxであるものを全て取得しましょう。
# pandasのread_excelメソッドを用い、'branch_offices'フォルダにある全てのxlsxファイルを読み込みましょう。
# そして、一つのデータフレームに統合しましょう。
# 3月で降順に並び替えましょう。
# applyや関数を用いて、上位5名の備考欄に賞与ありと記述しましょう。
# データフレームを"3月降順社員別売上データ"というファイル名でExcelファイルを出力しましょう。引数でindexはFalseに指定しておきましょう。
# osモジュールのrenameメソッドを使い、'branch_offices'フォルダにある全てのファイルに対して、"処理済み"という文言をファイル名に付け加えましょう。
```

```python
# globモジュールを使い'branch_offices'フォルダにある、拡張子がxlsxであるものを全て取得しましょう。
glob.glob('/content/sample_data/branch_office/*.xlsx')
```

```python
# pandasのread_excelメソッドを用い、'branch_offices'フォルダにある全てのxlsxファイルを読み込みましょう。
# そして、一つのデータフレームに統合しましょう。
df_02 = pd.DataFrame()
for i in glob.glob('/content/sample_data/branch_office/*.xlsx'):
  df = pd.read_excel(i)
  df_02 = pd.concat([df_02, df])
df_02.head(10)
```

```python
# 3月で降順に並び替えましょう。
df_02.sort_values(by = '3月売上', ascending = False, inplace = True)
df_02.reset_index(drop = True, inplace = True)
df_02.head(10)
```

```python
# applyや関数を用いて、上位5名の備考欄に賞与ありと記述しましょう。
df_02['備考欄'] = df_02['備考欄'].astype(str)
```

```python
bonuses = df_02[:5]
no_bonuses = df_02[5:]
```

```python
bonuses['備考欄'] = bonuses['備考欄'].apply(lambda x: (x + '賞与あり').replace('nan', ""))
bonuses
```

```python
# pd.DataFrameのconcatメソッドでbonusesとno_bonusesを統合
df_02_01 = pd.concat([bonuses, no_bonuses])
df_02_01.head(10)
```

```python
# assignメソッドを使用してbonus欄を追加する
bonuses = bonuses.assign(bonus = lambda x: '賞与あり')
bonuses
```

```python
# 備考欄をdropする
bonuses.drop('備考欄', axis = 1, inplace = True)
bonuses
```

```python
# bonusを賞与欄に置き換える
bonuses.rename({'bonus' : '賞与欄'}, axis = 1, inplace = True)
bonuses
```

```python
df_02_02 = pd.concat([bonuses, no_bonuses])
df_02_02.head(10)
```

```python
# itertuples使用でDataFrameを一行づつタプル型で取り出せる
for row in df_02.itertuples():
  if row[7] >= 72000:
    index_num = row[0]
    df_02.loc[index_num, '備考欄'] = '賞与有り'
df_02.head(10)
```

```python
# データフレームを"3月降順社員別売上データ"というファイル名でExcelファイル出力。
# 引数でindexはFalseに指定
df_02.to_excel('/content/sample_data/branch_office/3月降順社員別売上データ.xlsx', index = False)
```

```python
# 問題３　mergeを使ってデータを結合しよう
#『sales_data.xlsx』と『第4四半期各社員売上データ.xlsx』を開きましょう。
#『第4四半期各社員売上データ.xlsx』のカラム名'社員番号'を'社員ID'にrenameしましょう。
df_sales = pd.read_excel('/content/sales_data.xlsx')
df_sales.head(10)
```

```python
df_sales1 = pd.read_excel('/content/第4四半期各社員売上データ.xlsx')
df_sales1.head(10)
```

```python
# 二つのデータフレームを'社員ID'でmergeしましょう。
df_sales1.rename(columns = {'社員番号' : '社員ID'}, inplace = True)
df_sales1.head(10)
```

```python
submit_df = pd.merge(df_sales, df_sales1, how = 'left', on = '社員ID', suffixes = ['', '_重複'])
submit_df.head(10)
```

```python
# mergeしたデータから余分な列を削除 列名は['社員ID', '氏名','1月売上', '2月売上', '3月売上', '備考欄']だけにし、reindex
submit_df.drop(columns = ['性別',	'所属支店',	'1月売上_重複',	'2月売上_重複', '3月売上_重複',	'備考欄_重複'], inplace = True)
submit_df.reindex(columns = ['社員ID', '氏名','1月売上', '2月売上', '3月売上', '備考欄', '第4四半期売上'])
```

```python
# データフレームを"氏名入り売上データ"というExcelファイルを出力しましょう。引数でindexはFalseに指定しておきましょう。
submit_df.to_excel('/content/氏名入り売上データ.xlsx', index = False)
```
